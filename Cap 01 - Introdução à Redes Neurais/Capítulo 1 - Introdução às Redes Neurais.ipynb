# Um Breve Histrico

Desde o advento dos computadores, cientistas vem formulando maneiras de que as mquinas sejam capazes de receber entradas e retornar sadas desejadas para tarefas, por exemplo, de **classificao** e **regresso**. Alm disso, em aprendizado de mquina h o aprendizado **supervisionado** e o **no supervisionado**. Aprendizado de mquina supervisionado  usado quando temos os dados pr-estabelecios e rotulados que sero usados para o treinamento. Suponhamos que voc possui dados de sensores de um servidor que mede a taxa de uploads/downloads, temperatura e umidade a cada 10 minutos. Normalmente, esse servidor opera como o esperado e sem interrupes, porm algumas partes falham e causam uma pausa no servidor. Ns podemos coletar os dados e ento divid-los em duas classes: uma classe para tempo/observaes quando o servidor estava em operao normal e a outra classe para tempo/observaes quando o servidor estava parado. Quando o servidor est falhando, queremos o rotular os dados do sensor que levaram  falha, bem como os dados que precederam uma falha. Quando o servidor est operando normalmente, ns simplesmente rotularemos esses dados como "normal".

O que cada sensor mede nesse exemplo  chamado de *feature*. Um grupo de features compem um conjunto de features (representado como vetores/matrizes) e o valor de um conjunto de features pode ser referido como uma amostra. As amostras so dadas aos modelos de redes neurais para trein-las e que as mesmas retornem um resultado desejado a partir das entradas das amostras ou que prevejam algo durante a fase de inferncia.

os rtulos "normal" e "falha" so **classificaes** ou **rtulos**. Durante o seu estudo, voc ver essas informaes como **targets** ou **ground-truth**. Esses targets so as classificaes que so a *meta* ou o *alvo*, sabidamente  *verdadeiras e corretas*, para o algoritmo aprender. Para esse exemplo, o foco do treino do algoritmo  ler o dado do sensor e predizer se uma falha  eminente. Isto  apenas um exemplo de aprendizado supervisionado para classificao. Para a regresso, os dados so usados para prever valores numricos, como o preo de aes ou o aumento do combustvel. H tambm o aprenizado no supervisionado, onde o computador encontra estrutura nos dados sem conhecer os rtulos/classes antecipadamente. H conceitos adicionais (como aprendizado por reforo e aprendizado semi-supervisionado), mas o foco desse estudo  a classificao e regresso com redes neurais, mas o que veremos aqui pode ser usado em outros casos.

As *camadas densas*, a mais comum das camadas, consistem em interconectar os neurnios. Em uma camada densa, cada neurnio de uma determinada camada est conectado a cada neurnio da prxima camada, o que significa que o valor de sada de um neurnio  o valor de entrada do neurnio da camada seguinte. Cada conexo entre os neurnio possui um peso associado a ele, que um fator de treinamento de quanto dessa entrada usar e esse peso  multiplicado pelo valor de entrada. Uma vez que as entradas e os pesos fluem para o neurnio, eles so somados e um *bias* (outro parmetro de treinamento)  adicionado. O propsito do bias  deslocar a sada positivamente ou negativamente, que frequentemente nos ajuda a mapear mais tipos de dados dinmicos do mundo real.

O conceitos de pesos e biases podem ser considerados como "botes" que podem ser ajustaros para o treinamento do modelo. Em redes neurais, ns frequentement temos milhares ou milhes desses parmetros ajustados pelo otimizador durante o treino. Voc pode estar se perguntando: "por que no ter somente biases ou apenas os pesos?" Biases e pesos so ambos parmetros ajustveis e ambos tm impacto nas sadas dos neurnios, mas de maneiras diferentes. Como os pesos so multiplicados, eles mudaro apenas a magnitude ou mudaro o sinal do nmero (positivo e negativo). A equao da sada do neurnio no  to diferente da funo afim:

- Funo afim: f(x) = ax + b


- Funo do neurnio: sada = entrada.peso + bias

Uma rede neural tpica possi mlhares ou at mesmo milhes de **parmetros ajustveis** (pesos e biases). Seguindo essa ideia, as redes neurais atuam como enormes funes com um vasto nmero de **parmetros**. O conceito de uma funo longa com milhes de variveis que pode solucionar um problema no  muito difcil. Com muitas variveis relacionadas aos neurnios, dispostos em camadas interconectadas, podemos imaginar que existe diversas combinaes para essas variveis que retornaram as sadas desejadas. Encontrar a combinao dos parmetros (pesos e biases)  o verdadeiro desafio.

O objetivo final das redes neurais  ajustar os parmetros para que sejam aplicados em entradas desconhecidas, gerando uma sada desejada. Quando os algoritmos de aprendizado de mquina supervisionado so treinados, ns mostramos ao modelo exemplos de entrada e suas sadas respectivas. Um dos grandes problemas dessa tcnica  o **overfitting**. Esse problema acontece quando o modelo aprende somente com os dados de treino e no "entende" nada sobre as dependncias de entrada/sada subjacentes. Nesse caso, a rede neural apenas "memorizou" os dados de treino.

Desde modo, usamos os dados "da amostra" para treinar o modelo e os dados "de fora da amostra" para validar a rede neural. Certos padres so utilizados para auxiliar na criao do modelo, por exemplo, em um conjunto de dados de 100 mil amostrar (dados e rtulos), 10 mil desses dados sero separados para a etapa de validao, ou seja, o modelo treinar com 90 mil dados e ser avaliada com 10 mil dados, sendo esses ltimos dados que a rede neural nunca viu. O objetivo dessa etapa  qe o modelo no apenas preveja com uma boa precio os dados de treino, mas tamb seja capaz de prever com a mesma preciso os dados de validao que ele nunca viu.

Isso  chamado de **generalizao**, isto , o modelo aprendeu durante o treino ao invs de memorizar. A ideia  que "treinamos" (lento ajuste nos pesos e biases) uma rede neural com grandes quantidade de exemplos. Aps o treino, entregamos os dados que o modelo nunca viu para que o mesmo retorne o que desejamos com a mesma preciso do treino.

Agora voc deve ter um compreenso geral do que so as redes neurais, ou pelo menos qual  o objetivo e como planejamos atingir esse objetivo. Para treinar as redes neurais, ns calculamos o quo "errados" os algoritmos usados esto (chamado de **perda**) e tentar de modo lento o ajuste dos parmetros (pesos e biases) que ento, aps vrias iteraes, o modelo gradualmente erra menos. O objetivo de todas as redes neurais  generalizar, o que significa que a rede neural pode receber diversos exemplos nunca antes vistos e acertar com preciso a sada desejada. Redes neurais podem ser usadas mais do que apenas para problemas de classificao (esse tipo de problema  o mais comum do uso de redes naurais). Podemos us-las em problemas de regresso (prever um valor escalar), clusterizao (atribuir dados no estruturados em grupos) e muitas otras tarefas.
